```python
import pandas as pd
import psycopg2
```

```python
# Connect to PostgreSQL

conn = psycopg2.connect(
    host="localhost",
    database="bank_reviews",
    user="admin",
    password="admin123"
)
```

```python

def run_query(sql):
    return pd.read_sql(sql, conn)
```

```python
sql = "SELECT * FROM reviews;"

run_query(sql)
```

```python
# Count reviews per bank

sql = """
SELECT b.bank_name, COUNT(v.review_id) AS total_reviews
FROM banks b
LEFT JOIN reviews v ON b.bank_id = v.bank_id
GROUP BY b.bank_name
ORDER BY total_reviews DESC;
"""

run_query(sql)
```

```python
# Average rating per bank

sql = """
SELECT b.bank_name,
       ROUND(AVG(v.rating)::numeric, 2) AS avg_rating
FROM reviews v
JOIN banks b ON b.bank_id = v.bank_id
GROUP BY b.bank_name
ORDER BY avg_rating DESC;
"""

run_query(sql)
```
